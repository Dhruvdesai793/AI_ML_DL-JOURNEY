# Unsupervised ML
Unsupervised learning shares a property in common with supervised learning: it transforms
one dataset into another. But the dataset that it transforms into is not previously known or
understood. Unlike supervised learning, there is no “right answer” that you’re trying to get
the model to duplicate. You just tell an unsupervised algorithm to “find patterns in this data
and tell me about them.”


# Parametric vs. Nonparametric learning
## Oversimplified: Trial-and-error learning vs. counting and probability

### A parametric model is characterized by having a fixed number of parameters, whereas a nonparametric model’s number of parameters is infinite (determined by data).

As an example, let’s say the problem is to fit a square peg into the correct (square)
hole. Some humans (such as babies) just jam it into all the holes until it fits somewhere
(parametric). A teenager, however, may count the number of sides (four) and then search
for the hole with an equal number (nonparametric). Parametric models tend to use trial and
error, whereas nonparametric models tend to count.

#### Unsupervised learnnig is all about grouping data.


# Neural prediction: frwd propogation
#### WEIGHTS: In simple terms, a weight is a number that represents the strength or importance of a connection between two neurons. It is also a **measure of sensitivity**

````python
# Creating a basic neural network
# 1. Convert the split strings into floats immediately
l = [float(i) for i in input("enter the data separated by spaces: ").split()]
w = float(input("enter the weight for the network: "))

def neural_network(inputData, weights):
    prediction = []
    # 2. Iterate directly through the numeric values
    for i in inputData:
        # 3. Multiply the float input by the float weight
        prediction.append(i * weights) 
    return prediction

pred = neural_network(l, w)
print(pred)
````

### What is input data?
It’s a number that you recorded in the real world somewhere. It’s usually something
that is easily knowable, like today’s temperature, a baseball player’s batting average, or
yesterday’s stock price.
### What is a prediction ?
A prediction is what the neural network tells you, given the input data, such as “given the
temperature, it is 0% likely that people will wear sweatsuits today” or “given a baseball player’s batting average, he is 30% likely to hit a home run” or “given yesterday’s stock price, today’s stock price will be 101.52.”
### Is this prediction always right?
No. Sometimes a neural network will make mistakes, but it can learn from them. For example,
if it predicts too high, it will adjust its weight to predict lower next time, and vice versa.
### How does the network learn?
Trial and error! First, it tries to make a prediction . Then, it sees whether the prediction was too high or too low. Finally, it changes the weight (up or down) to predict more accurately the next time it sees the same input 

## Making a prediction with multiple inputs
> A relevant reminder:
*The interface for the neural network is simple: it accepts an input variable as information and a weights variable as knowledge, and it outputs a prediction.*

````python
import numpy as np

def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i]*b[i])
    return output

def neural_network(input_data,weights):
    pred = w_sum(input_data, weights)
    return pred
    
# The NumPy way
def neural_network(input_data, weights):
    return input_data.dot(weights) #replaces w_sum function


# 1. The knowledge(weights)
weights = np.array([0.2,0.1,0.0])

# 2. data
toes = [8.5, 9.5,9.9,9.0]
wlrec = [0.65,0.8,0.8,0.9]
nfans = [1.2,1.3,0.5,1.0]

# 3. selecting the input for the FIRST game (index0)
input_for_game = np.array([toes[0],wlrec[0],nfans[0]])

# 4. getting the prediction
pred = neural_network(input_for_game,weights)

print(f"Input: {input_for_game}")
print(f"Prediction: {pred}")
````

### Dot Product = Similarity: It measures how well the input "aligns" with the weights.

### The AND/OR Rule: The dot product checks if Input[0] AND Weight[0] exist. It does this for every position and then uses an OR to combine them.

####    Example: weights = [1, 0, 1] means "Give a high score if input[0] OR input[2] is present."

### The NOT Rule: A negative weight acts as a "NOT" gate.

####    Example: weights = [1, 0, -1] means "Give a high score if input[0] is present OR if input[2] is NOT present."

### The Scale Rule: A small weight (like 0.5) means the input must be twice as strong to have the same impact as a weight of 1.0.

````python
# numpy version of above code

import numpy as np
weights = np.array([0.1,0.2,0])
def neural_network(input,weights):
    pred = input.dot(weights)
    return pred

toes = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])
input = np.array([toes[0],wlrec[0],nfans[0]])

pred = neural_network(input,weights)
print(pred)
````

# Gradient Descent
### comparing gives a measurement of how much a prediction ***missed*** by. It is the 2nd step Predict.
### Learning tells each weight how it can change to reduce the Error.

### Squared Error method: 
* #### by squaring the difference b/w **prediction and the goal** (prioritizes large errors):
  * large erros get much bigger: difference is 10, err = 100
  * small errors get much smaller: diff is 0.01 , err = 0,0001
* #### forces +ve values:
  * The Problem: without squaring, a miss of +1000 and a miss of -1000 would average to 0 err, making the network look perfect when it actually failed 2x.
  * The solution: Squaring ensures every mistake adds to the "total error" pile, giving an honest view of performance
* #### Smoothness for math:
  * squaring creates a *"bowl-shaped"* curve (a parabola). This shape is crucial for **GRADIENT DESCENT** because it provides a clear, smooth slope.

### HOT AND COLD LEARNING:
* inefficient
* sometimes impossible to predict the exact goal prediction. 

````python
# Hot and Cold learning

weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001 # weight modifier for errs

for iteration in range(10): #higher reps, lower err

    prediction = input*weight
    error = (prediction - goal_prediction)**2

    print("Error:" + str(error) + " Prediction:" + str(prediction))

    up_prediction = input* (weight + step_amount)
    up_error = (goal_prediction - up_prediction)**2

    down_prediction = input*(weight-step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount

    if (down_error > up_error):
        weight = weight + step_amount
````

## Gradient Descent (code below)

1. Stopping:
 *  The Logic: If the error is zero, the learning should stop.
 *  How it works: Look at (pred - goal_pred). If your prediction is exactly equal to the goal, this part of the equation becomes 0. When you multiply by 0, the direction_and_amount becomes 0, and the weight stops changing. The network has "arrived" at the correct answer.

2. Negative Reversal:
 * The Logic: If you increase the weight, does the error go up or down?
 * How it works: This is handled by the input. If the input is positive, increasing the weight increases the prediction. But if the input is negative, increasing the weight would actually decrease the prediction. Multiplying by the input ensures that the weight moves in the correct direction to reduce error, regardless of whether the input data is positive or negative.

3. Scaling:
 * The Logic: If the input is huge, don't let the weight update explode.
 * How it works: Multiplying by input also acts as a volume knob. If the input is very large (e.g., 100), the "signal" for the error is amplified. If the input is tiny (e.g., 0.01), the weight update is tiny. It ensures that the weight doesn't overreact to a single piece of data if that data point is small and insignificant.

### Relationship
* Error: (pred−goal_pred)**2

* Delta (Pure Error): pred−goal_pred (Tells you the distance and direction).

* Direction and Amount (The Gradient): delta×input (This is the "Raw Gradient").

* Weight Update: weight=weight−(alpha×gradient)

* **Alpha** is the learning rate, u don't want your code to undershoot or overshoot

````python
# V1 primitive gradient descent
weight = 0.5
goal_pred = 0.8
input = 0.5
for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input #scaling, -ve reversal, and stopping
    weight = weight - direction_and_amount
    print("Error:" + str(error) + " Prediction:" + str(pred))
````

````python
# V2 gradient descent

weight = 0.5 
goal_pred = 0.8
input = 0.5
alpha = 0.1

for i in range(20):
    pred = input * weight
    error = (pred - goal_pred) **2

    delta = (pred - goal_pred)
    weight_delta = input * delta
    weight = weight - (weight_delta*alpha)

    print(f"Iteration: {i+1} | Error: {error:.10f} | Prediction: {pred:.10f}")
````
